```python
%%writefile app.py
import streamlit as st
from google import genai

# ----------------------------
# Gemini API
# ----------------------------

API_KEY = "YOUR API KEY"

client = genai.Client(api_key=API_KEY)

# ----------------------------
# Streamlit Page
# ----------------------------

st.set_page_config(
    page_title="AI Learning Buddy",
    page_icon="🤖",
    layout="centered"
)

st.title("🤖 AI Learning Buddy")

st.write("Learn any topic instantly using Google's Gemini AI.")

# ----------------------------
# User Input
# ----------------------------

topic = st.text_input(
    "Enter a topic",
    placeholder="Example: Machine Learning"
)

option = st.selectbox(
    "Choose an option",
    [
        "Explanation",
        "Real-Life Example",
        "Summary",
        "Quiz",
        "Interview Questions",
        "Fun Facts"
    ]
)

# ----------------------------
# Prompt Builder
# ----------------------------

def build_prompt(topic, option):

    if option == "Explanation":
        return f"Explain {topic} in simple language suitable for beginners."

    elif option == "Real-Life Example":
        return f"Give three real-life examples of {topic}."

    elif option == "Summary":
        return f"Summarize {topic} into easy bullet points."

    elif option == "Quiz":
        return f"Create a 5-question multiple-choice quiz on {topic} with answers."

    elif option == "Interview Questions":
        return f"Generate the top 10 interview questions on {topic} with answers."

    elif option == "Fun Facts":
        return f"Give 10 interesting facts about {topic}."

# ----------------------------
# Generate
# ----------------------------

if st.button("Generate"):

    if topic.strip() == "":
        st.warning("Please enter a topic.")

    else:

        prompt = build_prompt(topic, option)

        with st.spinner("Generating..."):

            response = client.models.generate_content(
                model="gemini-2.5-flash",
                contents=prompt
            )

        st.success("Completed!")

        st.markdown("## Result")

        st.write(response.text)

st.markdown("---")
st.caption("Made using Streamlit + Gemini AI")
```
